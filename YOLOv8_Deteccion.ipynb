# üîç Detecci√≥n de Objetos con YOLOv8 en Google Colab

Este notebook permite probar la detecci√≥n de objetos usando el modelo YOLOv8 con la librer√≠a **Ultralytics**.


```python
# üìå Paso 1: Instalaci√≥n de librer√≠as
!pip install ultralytics opencv-python --quiet
```

```python
# üìå Paso 2: Importar librer√≠as necesarias
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
# üìå Paso 3: Cargar modelo YOLOv8 pre-entrenado
model = YOLO('yolov8n.pt')  # n = nano (m√°s r√°pido y liviano)
```

```python
# üìå Paso 4: Subir una imagen para detecci√≥n
from google.colab import files
uploaded = files.upload()
```

```python
# üìå Paso 5: Leer imagen subida (por ejemplo, perro.jpg)
import io
import numpy as np
from PIL import Image

img_path = next(iter(uploaded))  # nombre del archivo subido
img = cv2.imread(img_path)
```

```python
# üìå Paso 6: Realizar detecci√≥n
results = model(img)
```

```python
# üìå Paso 7: Mostrar resultados
results[0].show()  # Muestra con interfaz gr√°fica de Ultralytics

# Alternativamente, mostrar con OpenCV en Colab
annotated_img = results[0].plot()
cv2_imshow(annotated_img)
```
