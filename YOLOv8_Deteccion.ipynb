# 🔍 Detección de Objetos con YOLOv8 en Google Colab

Este notebook permite probar la detección de objetos usando el modelo YOLOv8 con la librería **Ultralytics**.


```python
# 📌 Paso 1: Instalación de librerías
!pip install ultralytics opencv-python --quiet
```

```python
# 📌 Paso 2: Importar librerías necesarias
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
# 📌 Paso 3: Cargar modelo YOLOv8 pre-entrenado
model = YOLO('yolov8n.pt')  # n = nano (más rápido y liviano)
```

```python
# 📌 Paso 4: Subir una imagen para detección
from google.colab import files
uploaded = files.upload()
```

```python
# 📌 Paso 5: Leer imagen subida (por ejemplo, perro.jpg)
import io
import numpy as np
from PIL import Image

img_path = next(iter(uploaded))  # nombre del archivo subido
img = cv2.imread(img_path)
```

```python
# 📌 Paso 6: Realizar detección
results = model(img)
```

```python
# 📌 Paso 7: Mostrar resultados
results[0].show()  # Muestra con interfaz gráfica de Ultralytics

# Alternativamente, mostrar con OpenCV en Colab
annotated_img = results[0].plot()
cv2_imshow(annotated_img)
```

## 📎 Notas importantes (Breakdown)

- ✅ **Modelo usado**: `yolov8n.pt` es el más liviano (nano), ideal para pruebas rápidas.
- 🔁 Puedes cambiar a otros modelos disponibles según el balance entre velocidad y precisión:
  - `yolov8s.pt` → small
  - `yolov8m.pt` → medium
  - `yolov8l.pt` → large
  - `yolov8x.pt` → extra-large
- 📷 **Subida de imagen**: utiliza `files.upload()` para subir imágenes desde tu equipo.
- 👁️ **Visualización en Colab**: usa `cv2_imshow()` (no `cv2.imshow`) para mostrar imágenes correctamente en Google Colab.
- 🧠 **Detección sin entrenamiento**: este ejemplo utiliza un modelo pre-entrenado en COCO (80 clases).
